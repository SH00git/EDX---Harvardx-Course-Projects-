```python
import pandas as pd

whisky =pd.read_csv("whisky.txt")
whisky['region'] =pd.read_csv("regions.txt")
print(whisky.columns)
whisky.iloc[:10,:]
```

```python
flavour = whisky.iloc[:,2:-4]
flavour
```

```python
corr_flavour = pd.DataFrame.corr(flavour)
corr_flavour
```

## Correlation btw whiskey flavours 

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.pcolor(corr_flavour)
plt.colorbar()

plt.savefig("flavour_corr.pdf")
```

## Correlation between whiskey distillaries  

```python
corr_whiskey = pd.DataFrame.corr(flavour.transpose())
corr_whiskey 
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.pcolor(corr_whiskey)
plt.colorbar()

plt.savefig("whisy_corr.pdf")
```
